```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.formula.api as sm
```

## plotting tables and regression charts for varying `k` variable in KDTree and QuadTree

```python

folder_path = '../'
data_frames = []

varying_k_files = ['varyingkUniformKDTree.csv', 'varyingkUniformQuadTree.csv', 'varyingkGaussianKDTree.csv', 'varyingkGaussianQuadTree.csv']

for file_name in varying_k_files:
   file_path = os.path.join(folder_path, file_name)
   if os.path.exists(file_path):
      df = pd.read_csv(file_path)
      data_frames.append(df)
   else:
       print("path not found")

vk_uk_df = data_frames[0]
vk_uq_df = data_frames[1]
vk_gk_df = data_frames[2]
vk_gq_df = data_frames[3]

plt.figure(figsize=(10, 5))

# Plot each DataFrame
plt.plot(vk_gq_df.iloc[:, 0], vk_gq_df.iloc[:, 1], marker='o', linestyle='-', label='Varying k Gaussian Quad Tree')
plt.plot(vk_uq_df.iloc[:, 0], vk_uq_df.iloc[:, 1], marker='o', linestyle='-', label='Varying k Uniform Quad Tree')
plt.plot(vk_gk_df.iloc[:, 0], vk_gk_df.iloc[:, 1], marker='o', linestyle='-', label='Varying k Gaussian KD Tree')
plt.plot(vk_uk_df.iloc[:, 0], vk_uk_df.iloc[:, 1], marker='o', linestyle='-', label='Varying k Uniform KD Tree')

plt.title('Comparison of KD Tree and Quad Tree KNNQuery performance with a varying `k` value')
plt.xlabel('k(number of nearest neighbors)')  # Replace with actual label
plt.ylabel('time(micro-seconds)')  # Replace with actual label
plt.legend()
plt.grid(True)
plt.show()

```

```python
#regression for this first table
#testing the varying k on uniform points in the kd tree
model_vk_uk_uq = sm.ols(formula='avgTime ~ k', data = vk_uk_df).fit()
print(model_vk_uk_uq.summary())
```

# timing the buckets -------------------

```python
folder_path = '../'
data_frames = []

column_names = ['k', 'avgTime']
bucket_files = ['varyingkGaussianBucket.csv','varyingkUniformBucket.csv']


for file_name in bucket_files:
   file_path = os.path.join(folder_path, file_name)
   if os.path.exists(file_path):
      df = pd.read_csv(file_path, header=None, names = column_names)
      data_frames.append(df)
   else:
       print("path not found")

bG_df = data_frames[0]
bU_df = data_frames[1]

plt.figure(figsize=(10, 5))

# Plot each DataFrame
plt.plot(bG_df.iloc[:, 0], bG_df.iloc[:, 1], marker='o', linestyle='-', label='Varying k bucket gaussian distribution')
plt.plot(bU_df.iloc[:, 0], bU_df.iloc[:, 1], marker='o', linestyle='-', label='Varying k bucket uniform distribution')


plt.title('Comparison of the provided Bucket class for uniform and gaussian distribution with varying (k)')
plt.xlabel('k(number of nearest neighbors)')  # Replace with actual label
plt.ylabel('time(micro-seconds)')  # Replace with actual label
plt.legend()
plt.grid(True)
plt.show()
```

```python
bG_df['distribution'] = 'Gaussian'
bU_df['distribution'] = 'Uniform'

combined_df = pd.concat([bG_df, bU_df], ignore_index = True)
combined_df.columns = ['k', 'avgTime', 'distribution']
model = sm.ols(formula='avgTime ~ k * distribution', data=combined_df).fit()

print(model.summary())
```

## regression interperetation from bucket when comparing gaussian and uniform points
- it is seen that uniform distribution generally results in quicker average times compared to gaussian distribution
- this is seen in the intercept for uniform distribution is significantly lower than the gaussian distribution
- uniform distribution in a bucket tends to perform KNN queries quicker because the points are more spread out and the buckets do not get too full
